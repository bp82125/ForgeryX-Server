```python
import os
import cv2
import numpy as np

from photoholmes.utils.image import read_image, save_image, plot, create_heatmap
```

```python
img_folder = "/home/nhat82125/ForgeryX-Server/example"
img_name = "im1_edit3"

tp_img_path = os.path.join(img_folder, f"{img_name}.jpg")
gt_img_path = os.path.join(img_folder, f"{img_name}.png")

tp_img = read_image(tp_img_path)
gt_img = read_image(gt_img_path)

plot(tp_img, "Tampered Image")
plot(gt_img, "Ground Truth")
```

```python
from photoholmes.methods.wavelet import Wavelet, wavelet_preprocessing

image_data = {"image": tp_img}
image_input = wavelet_preprocessing(**image_data)

method = Wavelet()

output = method.predict(**image_input)
heatmap = create_heatmap(output)

plot(heatmap, "Wavelet Prediction")
save_image("wavelet.png", heatmap)
```
